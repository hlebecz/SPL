```python
# 1.7 Создать массив 10x10 со случайными значениями, найти минимум и максимум
random_array = np.random.rand(10, 10)
min_val = np.min(random_array)
max_val = np.max(random_array)
print(f"1.7 Минимум: {min_val}, Максимум: {max_val}")
```

```python
# 1.9 Создать 8x8 матрицу и заполнить её в шахматном порядке
chessboard = np.zeros((8, 8))
chessboard[1::2, ::2] = 1  # Нечетные строки, четные столбцы
chessboard[::2, 1::2] = 1  # Четные строки, нечетные столбцы
print("1.9 Шахматная доска 8x8:")
print(chessboard)
```

```python
# 1.11 Проверить, одинаковы ли 2 numpy массива
arr1 = np.array([1, 2, 3])
arr2 = np.array([1, 2, 3])
are_equal = np.array_equal(arr1, arr2)
print(f"1.11 Массивы одинаковы: {are_equal}")
```

```python
# 1.12 Заменить максимальный элемент на ноль
arr = np.array([1, 5, 3, 9, 2])
arr[np.argmax(arr)] = 0
print("1.12 Массив после замены максимума на 0:", arr)
```

```python
# 1.13 Найти наиболее частое значение в массиве
arr = np.array([1, 2, 2, 3, 3, 3, 4])
unique, counts = np.unique(arr, return_counts=True)
most_frequent = unique[np.argmax(counts)]
print(f"1.13 Наиболее частое значение: {most_frequent}")
```

```python
# 1.14 Найти n наибольших значений в массиве
arr = np.array([1, 5, 3, 9, 2, 7])
n = 3
largest_values = np.partition(arr, -n)[-n:]
print(f"1.14 {n} наибольших значений: {largest_values}")
```

```python
# === 2.1 Импортировать датасет ===
df = pd.read_csv("test.csv")

# Посмотрим первые строки
print("Первые строки датасета:")
print(df.head())
```

```python
# === 2.2 Взять 1000 значений из выбранного датасета ===
df_sample = df.sample(1000, random_state=42)
```

```python
# === 2.3 Проверить данные на пропуски ===
print("\nКоличество пропусков по столбцам:")
print(df_sample.isnull().sum())
```

```python
# === 2.4 Заполнить пропуски и обработать аномальные значения ===
# Пример: числовые пропуски — заполняем медианой, категориальные — модой
for col in df_sample.columns:
    if df_sample[col].dtype == 'object':
        df_sample[col] = df_sample[col].fillna(df_sample[col].mode()[0])
    else:
        df_sample[col] = df_sample[col].fillna(df_sample[col].median())
```

```python
# Удаляем выбросы (метод межквартильного размаха)
numeric_cols = df_sample.select_dtypes(include=['float64', 'int64']).columns
for col in numeric_cols:
    Q1 = df_sample[col].quantile(0.25)
    Q3 = df_sample[col].quantile(0.75)
    IQR = Q3 - Q1
    df_sample = df_sample[~((df_sample[col] < (Q1 - 1.5 * IQR)) | (df_sample[col] > (Q3 + 1.5 * IQR)))]
```

```python
print(df_sample.dtypes)

```

```python
# === 2.5 Провести полный анализ ===
# Примерный анализ распределения данных
print("\nОсновная статистика числовых признаков:")
print(df_sample.describe())

# Гистограммы
df_sample.hist(bins=30, figsize=(15, 10))
plt.suptitle("Распределения признаков после очистки")
plt.show()

# Корреляционная матрица
plt.figure(figsize=(10, 6))
plt.title("Корреляционная матрица")
numeric_cols = df_sample.select_dtypes(include=['float64', 'int64'])
plt.imshow(numeric_cols.corr(), cmap='coolwarm', interpolation='none')
plt.colorbar(label='Корреляция')
plt.xticks(range(len(numeric_cols.corr().columns)), numeric_cols.corr().columns, rotation=90)
plt.yticks(range(len(numeric_cols.corr().columns)), numeric_cols.corr().columns)
plt.show()
```

```python
# === 2.6 Сохранить итоговый массив без выбросов и пропусков ===
df_sample.to_csv("surname.csv", index=False)
print("\nФайл 'surname.csv' сохранён без выбросов и пропусков.")
```
